1) Using the data from the hourly time series at Deep Cove (dc), show that the canned fft routine returns the correct discrete Fourier Transform of a suitable subset of the data (your choice on what subset). You can do this graphically, by comparing the amplitude and the phase of the Fourier components, or simply show that the resulting numbers are the same.
Some Hints:
ignore bad data (rather than dealing with it "properly"). Make sure the units of your FT are: oCs.
If you want to compute the DFT efficiently, form the matrix F referred to in the notes, and plot some of the time series associated with the different entries and make sure they look like Sines and Cosines and that the edge conditions are what you want. Alternately, you can use your dft routine from the last assignment, and chose your frequencies properly.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]

dc= dc[~np.isnan(dc)] #to fix nan error 


def dft(x,dt,f):
    t = dt*np.arange(0,len(x))
    T,F=np.meshgrid(t,f)
    # make the fourier matrices
    E = np.exp(-1j*2*np.pi*T*F)
    X = dt*np.dot(E,x)
    return X

N = 720   #30 days worth of data
T= 720.*3600.
fNover2= 1./7200.  #nyquist frequency with delta t=3600s. dt will be 3600s for this case b/c data is hourly

x= dc[0:N:]
f= np.linspace(0,fNover2,N/2+1)

Xone = dft(x,3600.,f)

Xtwo =(T/N)*np.fft.fft(x)
Xtwo = Xtwo[0:360:]        #[360:720:] mirrors first half of data

print('for Xone')
print np.shape(Xone)
print Xone[:6]
print Xone[-5:]

print('for Xtwo')
print np.shape(Xtwo)
print Xtwo[:6]
print Xtwo[-5:]

fig,ax=plt.subplots(1,1,)                                 #can see that Xone = Xtwo , although Xone has one extra entry.
ax.plot(Xone[:100:], label='Xone', linestyle='--')                        #did only 100/360 data points for better graph view
ax.plot(Xtwo[:100:], label='Xtwo',marker='*')
ax.set_xlabel('data point');ax.set_ylabel('X [C^2s^2]')
ax.legend()
```

1) Write a short function to create a "red" time series by taking the cumulative sum of Normally distributed random data, and plot one realization. The number of data points N can be relatively small (say 1000).

```python
N= 1000
red = np.cumsum(np.random.randn(N))
t = np.arange(N)
fig,ax=plt.subplots(1,1)
ax.plot(t,red)
ax.set_xlabel('t [s]')
ax.set_ylabel('red [units]')

```

2) Write a second function to return the estimate of the power spectrum of x. For posterity sake, include a value of the sampling frequency, and return the proper frequencies for the estimate. Plot the resulting spectrum

```python
#proved the canned Gxx routine worked in question 1-part2 so will use for power spectrum
dt= 1
N=1000
Gxx,f=mlab.psd(red,Fs=1.,NFFT=N) #where the sampling frequency is equal to 1Hz because dt is 1s.
nyquist = 1./2.*dt 
f= np.linspace(0, nyquist, N/2+1)

fig,ax=plt.subplots(1,1)
ax.loglog(f[1:],Gxx[1:])
ax.set_xlabel('f [Hz]');ax.set_ylabel('$|G_{xx}(f)| [units^{2} Hz^{-1}]$');
ax.set_title('power spectrum as a function of frequency')
```

3) Now do the Monte-Carlo, by running the above 100 times and plotting the resulting spectral estimates as dots on the same plot. As you are making your data, save it into an array.




```python
import matplotlib.pyplot as plt

frq= np.zeros((100,501))
Gxs = np.zeros((100,501))

fig = plt.figure()
ax=fig.add_subplot(1,1,1)
N=1000
Gxx,f=mlab.psd(red,Fs=1.,NFFT=N)

for i in range(0,100): 
    h = np.random.randn(N)
    red = np.cumsum(h)
    Gxx,f=mlab.psd(red,Fs=1.,NFFT=N)
    Gxs[i,:]= Gxx
    frq[i,:]= f
    ax.plot(frq[i,:],Gxs[i,:], '.')
ax.loglog()
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('Gxx(f)')
ax.set_title('100 iterations of Gxx vs f')
```

4) Now, on the same plot, add the mean of all your Gxx(f) estimates (as a function of f), and the upper and lower bounds encompasing 95% of the data. (Hint, sort each column of your matrix - np.sort - and plot the value at 0.025N and 0.975N).

5) Show that the ratio of the lower and upper bounds of your spectral estimates spread compares favourably with the ratio between the lower and upper 95% confidence bounds of the Ï‡22 distribution. Comment on any difference.

```python
import matplotlib.pyplot as plt
import scipy.stats as stats

frq= np.zeros((100,501))
Gxs = np.zeros((100,501))

fig = plt.figure()
ax=fig.add_subplot(1,1,1)
N=1000
Gxx,f=mlab.psd(red,Fs=1.,NFFT=N)



for i in range(0,100): 
    h = np.random.randn(N)
    red = np.cumsum(h)
    Gxx,f=mlab.psd(red,Fs=1.,NFFT=N)
    Gxs[i,:]= Gxx
    frq[i,:]= f 
    
   
    ax.loglog(frq[i,:],Gxs[i,:], '.')
    inter = stats.chi2.interval(0.95,df=2)
    ax.fill_between(frq[i,:],2.*Gxs[i,:]/inter[1],2.*Gxs[i,:]/inter[0],alpha=0.5,      
                linewidth=0.0,edgecolor=None,color=None,facecolor='y')              # confidence bound shading encompasses 
                                                                                    ## 95% of data (yellow)

   
Gxxmean = np.mean(Gxs, axis = 0)
ax.loglog(f,Gxxmean, 'b-', label = 'mean')      #Gxx mean is blue line 
upper = Gxxmean + 2*np.std(Gxs, axis = 0)
lower = Gxxmean - 2*np.std(Gxs, axis = 0)           #attempts at upper and lower boundaries


ax.loglog(f,upper, 'g-', label= 'upper bound 97.5%')   # upper/lower bounds are not the same as the 95% confidence bounds
ax.loglog(f,lower, 'r-', label = 'lower bound 2.5%')   

ax.legend()
ax.set_xlabel('frequency (Hz)')
ax.set_ylabel('Gxx(f)')
ax.set_title('100 iterations of Gxx and Gxxmean vs f with error bounds')
```

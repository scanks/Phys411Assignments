```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib nbagg

hourdata=np.genfromtxt('http://web.uvic.ca/~jklymak/Phy411/Data/AllHourly.txt')[[6,28],2:]
dc = hourdata[0,:]
jb=hourdata[1,:]

dc= dc[~np.isnan(dc)] #to fix nan error 


def dft(x,dt,f):
    t = dt*np.arange(0,len(x))
    T,F=np.meshgrid(t,f)
    # make the fourier matrices
    E = np.exp(-1j*2*np.pi*T*F)
    X = dt*np.dot(E,x)
    return X

N = 720   #30 days worth of data
fNover2= 1./7200.  #nyquist frequency with delta t=3600s. dt will be 3600s for this case b/c data is hourly
x= dc[0:N:]
f= np.linspace(0,fNover2,N/2)

Xone = dft(x,3600.,f)

Xtwo = np.fft.fft(x)

print('for Xone')
print np.shape(Xone)
print Xone[:6]
print Xone[-5:]
print 'Xone[N/2]= %1.3e %+1.3ei '%(Xone[180].real,Xone[180].imag)

print('for Xtwo')
print np.shape(Xtwo)
print Xtwo[:6]
print Xtwo[-5:]
print 'Xtwo[N/2]= %1.3e %+1.3ei '%(Xtwo[180].real,Xtwo[180].imag)

```
